## 2. Dynamic Prompt Engineering with Templates

Prompt templates allow dynamic input injection and help maintain consistency in prompt design across multiple use cases.

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

print(template.format(topic="LangChain"))
```

## LLM Initialization (Setup)

## 5. Tool Integration and Agent Capabilities

Tools allow LangChain to interact with external systems. Here we demonstrate a simple calculator tool.

```python
from langchain_core.tools import Tool

def calculator(input):
    return eval(input)

tool = Tool(
    name="Calculator",
    func=calculator,
    description="Performs basic math calculations"
)

print(tool.invoke("5+3"))
```
