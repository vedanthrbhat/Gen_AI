```python
from langchain_core.prompts import PromptTemplate
from langchain_community.chat_models import ChatOllama


llm = ChatOllama(
    model="llama3",
    temperature=0.3
)
```

```python
code_review_template = PromptTemplate(
    input_variables=["language", "code", "focus"],
    template="""
You are a senior software engineer performing a professional code review.

Language: {language}

Focus Areas:
{focus}

Code:
{code}

Tasks:
1. Identify bugs or logical issues
2. Suggest improvements (performance, readability, best practices)
3. Highlight security concerns if any
4. Provide a cleaned-up version of the code

Output Format:
- Issues:
- Improvements:
- Security:
- Refactored Code:
"""
)
```

```python
def review_code(language, code, focus):
    
    prompt = code_review_template.format(
        language=language,
        code=code,
        focus=focus
    )
    
    response = llm.invoke(prompt)
    
    return response.content
```

```python
sample_code = """
def add(a,b):
 return a+b
"""

result = review_code(
    language="Python",
    code=sample_code,
    focus="bugs, readability, best practices"
)

print(result)
```

```python
chain = code_review_template | llm

response = chain.invoke({
    "language": "Python",
    "code": "def add(a,b): return a+b",
    "focus": "performance, readability"
})

print(response.content)
```
